# Question 03: [Marks 5]

## Discuss the vanishing gradient problems in deep neural networks. Explain why they occur during backpropagation, how activation functions contribute to this issue, and suggest at least three techniques to overcome these problems.


## Write Answer 03:
Vanishing gradient problem occurs in deep neural networks when gradients become very small during backpropagation. Because of this, early layers learn very slowly or sometimes do not learn properly at all.

It happens because backpropagation uses the chain rule where gradients are multiplied layer by layer. If these values < 1 , repeated multiplication through many layers makes the gradient more small and more until it becomes almost 0.

Activation functions like sigmoid and tanh also add to this problem because their derivatives are small. So when gradients pass through many layers they keep getting reduced.

Due to this training becomes slow, early layers stop learning properly and overall performance goes down.

To reduce this problem ReLU activation is used because it does not make gradients smaller in same way. Better weight initialization like Xavier or He also helps to keep values stable. Batch normalization also helps by keeping values in normal range during training.

# **Artificial Neural Network vs Convolutional Neural Network**

**Structure**

ANN is made of fully connected layers where each neuron is connected to all neurons of the next layer. It usually has input layer, hidden layers and output layer.

CNN has a different structure. It contains convolution layers, pooling layers, activation layers and finally fully connected layers. It uses filters and local connections instead of full connections.

**How it works**

ANN takes input as a flat vector and each neuron calculates weighted sum then applies activation function. It does not understand position based relationships in data.

CNN works on data where position and arrangement of values matters like images where nearby pixels are related and form shapes and patterns. It uses filters to scan input, extract features and pooling reduces size while keeping important information.

**Feature extraction**

ANN does not automatically extract features. Features often need to be manually prepared before input.

CNN automatically learns features step by step like edges, textures, shapes and complex patterns from raw data.

**Parameters and computation**

ANN has large number of parameters because every neuron is connected to every neuron in next layer. So it needs more computation and can overfit easily.

CNN uses weight sharing and local connections so number of parameters is less and computation is more efficient.

**Advantages**

ANN
Simple structure
Good for general problems
Easy to implement

CNN
Very good for image and spatial type data
Automatic feature learning
More efficient for high dimensional data

**Limitations**

ANN
Cannot capture position based relationships in data
Not good for image type data
High chance of overfitting

CNN
More complex architecture
Needs more computing power
Not useful for simple tabular data

**Use cases**

ANN is used for structured data like tables, simple prediction problems and basic classification tasks.

CNN is mainly used for image recognition, video processing and other data where position and arrangement matters.

Activation Functions in Neural Networks
# ReLU

Formula:
f(x) = max(0, x)

Advantages:

Very simple and fast to compute

Helps reduce vanishing gradient problem

Works well in deep networks

Faster training compared to sigmoid and tanh


Disadvantages:

Dying ReLU problem (some neurons stop learning)

Not zero centered output

Use cases:

Most commonly used in hidden layers of deep neural networks
CNN and deep learning models
# Sigmoid Function

Formula:
sigma(x) = 1 / (1 + e^-x)

Advantages:

Output is between 0 and 1

Smooth and easy to differentiate

Useful for probability based output

Disadvantages:

Vanishing gradient problem

Output not zero centered

Slow learning in deep networks

Use cases:

Output layer in binary classification problems
# Tanh Function

Formula:
tanh(x) = (e^x − e^-x) / (e^x + e^-x)

Advantages:

Output is zero centered (-1 to 1)

Better than sigmoid for hidden layers

Gives stronger gradients compared to sigmoid

Disadvantages:

Still suffers from vanishing gradient problem

Slower compared to ReLU

Use cases:

Hidden layers in older neural networks
# Why ReLU is widely used in deep learning

ReLU is widely used because it is simple, fast, and helps deep networks train efficiently. Unlike sigmoid and tanh, it does not compress values into a small range, so gradients do not become very small easily. This helps avoid vanishing gradient problems and allows faster learning. It also makes the network sparse, which improves efficiency and performance in deep models.

# Pooling in Convolutional Neural Networks (CNN)

Pooling is a technique used in CNN to reduce the size of feature maps. It takes a small region from the input and converts it into a single value. This helps to reduce data size while keeping important information.



## Types of Pooling

### Max Pooling

In max pooling, the maximum value is selected from a region.

Example (2x2 region):  
[1 3  
 2 4] takes 4   

**Advantages:**  
-Keeps strongest features  
-Good for detecting edges and patterns  
-Helps model focus on important parts  

**Impact on feature extraction:**  
-Preserves important features clearly  
-Ignores weaker or less useful values  



### Average Pooling

In average pooling, the average value of the region is taken.

Example (2x2 region):  
[1 3  
 2 4] takes (1+3+2+4)/4 = 2.5  

**Advantages:**  
-Produces smoother output  
-Considers all values  

**Impact on feature extraction:**  
-May lose strong features  
-Gives overall general information  


## Comparison between Max and Average Pooling

Max pooling:  
-Takes highest value  
-Keeps strong features  
-Better for feature detection  

Average pooling:  
-Takes average value  
-Smooth output  
-May reduce feature strength  



## Role of Pooling in CNN

-Reduces size of feature map  
-Decreases number of parameters  
-Makes computation faster   



## Pooling and Overfitting

Pooling helps reduce overfitting because:  
-It reduces model complexity  
-Removes unnecessary details  
-Makes model more general instead of memorizing data  



## Pooling and Computational Complexity

-Smaller feature maps - less computation  
-Faster processing  
-Less memory usage  



Pooling is an important step in CNN that reduces data size while keeping important features. Max pooling is mostly used because it keeps strong features, while average pooling gives smoother results. It helps improve efficiency and reduces overfitting.

## Design and implement a deep Convolutional Neural Network (CNN) for a multi-class image classification problem using PyTorch. The network should consist of four convolutional blocks, where each block includes two convolution layers followed by activation and normalization, and a pooling layer at the end. The number of filters should increase progressively across blocks. After the convolutional part, flatten the feature maps and connect them to fully connected layers with dropout for regularization, and produce final outputs corresponding to K classes. Also explain the role of each part of the network in detail.

# Write Answer 10:
# Explanation of CNN Architecture

## 1. Convolution Layers
-Extract features from input image  
-First layers detect simple patterns (edges, lines)  
-Deeper layers detect complex patterns (shapes, objects)  

## 2. Activation Function (ReLU)
-Adds nonlinearity  
-Helps model learn complex patterns  
-Makes training faster and stable  


## 3. Batch Normalization
-Normalizes output of each layer  
-Makes training faster and more stable  
-Reduces internal covariate shift  



## 4. Pooling Layer (MaxPooling)
-Reduces the size of feature maps

-Keeps important features  
-Reduces computation and overfitting  


## 5. Increasing Filters (32 -> 64 -> 128 -> 256)
-Early layers learn simple features  
-Deeper layers learn complex features  
-Improves representation power  

## 6. Flatten Layer
-Converts 2D feature maps into 1D vector  
-Required before fully connected layers  


## 7. Fully Connected Layers
-Combine all extracted features  
-Perform final classification  

## 8. Dropout
-Randomly disables neurons during training  
-Reduces overfitting  
-Improves generalization  

## Final Output Layer
-Produces output for K classes  
-Used for multiclass classification because Softmax applied during training

```python
import torch
import torch.nn as nn

class DeepCNN(nn.Module):
    def __init__(self, num_classes):
        super().__init__()

        self.features = nn.Sequential(

            #32 filters
            nn.Conv2d(3, 32, kernel_size=3, padding='same'),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.Conv2d(32, 32, kernel_size=3, padding='same'),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2),

            #64 filters
            nn.Conv2d(32, 64, kernel_size=3, padding='same'),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.Conv2d(64, 64, kernel_size=3, padding='same'),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2),

            #128 filters
            nn.Conv2d(64, 128, kernel_size=3, padding='same'),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.Conv2d(128, 128, kernel_size=3, padding='same'),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.MaxPool2d(2),

            #256 filters
            nn.Conv2d(128, 256, kernel_size=3, padding='same'),
            nn.BatchNorm2d(256),
            nn.ReLU(),
            nn.Conv2d(256, 256, kernel_size=3, padding='same'),
            nn.BatchNorm2d(256),
            nn.ReLU(),
            nn.MaxPool2d(2),
        )

        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(256 * 8 * 8, 512),   # lets assume 128x128 input image
            nn.ReLU(),
            nn.Dropout(0.5),

            nn.Linear(512, 256),
            nn.ReLU(),
            nn.Dropout(0.3),

            nn.Linear(256, num_classes)
        )

    def forward(self, x):
        x = self.features(x)
        x = self.classifier(x)
        return x
```
