```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv("/content/drive/MyDrive/Datasets/Original_Dataset.csv")
df
```

```python
df.fillna(0,inplace=True)
```

```python
column_values =df[['Symptom_1', 'Symptom_2', 'Symptom_3', 'Symptom_4',
       'Symptom_5', 'Symptom_6', 'Symptom_7', 'Symptom_8', 'Symptom_9',
       'Symptom_10', 'Symptom_11', 'Symptom_12', 'Symptom_13', 'Symptom_14',
       'Symptom_15', 'Symptom_16', 'Symptom_17']].values.ravel()
#ravel flattens 2D array into 1D array
print(column_values)
```

```python
column_values.shape
```

```python
symps = pd.unique(column_values).tolist()
symp = [i for i in symps if str(i) != "0"]#not counting 0 as unique...

print("The unique types of symptoms:",len(symp))
```

```python
disease_symptom_df = pd.DataFrame(columns=['Disease'] + symp)
disease_symptom_df.head()
```

```python
disease_symptom_df['Disease']=df['Disease']
disease_symptom_df.shape
```

```python
disease_symptom_df.head()
```

```python
symptoms_list = [] # Initialize an empty list to store symptoms for each row

for _, row in df.iterrows():
    # Get all symptom columns, excluding the 'Disease' column
    values = row[[col for col in df.columns if col.startswith('Symptom_')]].tolist()

    # Filter out '0' values from the symptom list
    clean_values = [s for s in values if s != 0]
    symptoms_list.append(clean_values)

# Assign the collected list of symptom lists to the new 'symptoms' column in df
df["symptoms"] = symptoms_list
```

```python
df.head()
```

```python
disease_symptom_df.iloc[:, 1:] = 0

# Iterate over each row in disease_symptom_df
for i in range(len(disease_symptom_df)):
    symptoms = df['symptoms'].iloc[i]  # Get the list of symptoms for the current row

    # Update the columns based on the presence of symptoms
    for symptom in symptoms:
        if symptom in symp:
            disease_symptom_df.at[i, symptom] = 1
```

```python
disease_symptom_df.head()
```

```python
disease_symptom_df.shape
```

```python
X = disease_symptom_df.drop(columns=['Disease'])
y = disease_symptom_df['Disease']

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define models
models = {
    "Logistic Regression": LogisticRegression(),
    "Random Forest": RandomForestClassifier(),
    "Support Vector Machine": SVC(),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Naive Bayes": GaussianNB()
}

# Train and evaluate each model
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"{name} Accuracy: {accuracy:.4f}")
```

```python
def normalize(s):
    return s.strip().lower()

# Get unique doctor specialities
unique_specialities = doctors['Speciality'].unique()

# Normalize both lists
normalized_speciality_set = set(normalize(s) for s in unique_specialities)
normalized_dict_set = set(normalize(s) for s in dictionary.values())

# Find matching specialists
matching_specialists = normalized_speciality_set.intersection(normalized_dict_set)

# Print matching specialists
for specialist in matching_specialists:
    print(specialist.title())
```

```python
df = doctors.copy()

# Qualification scoring
qualification_scores = {
    'MBBS': 1,
    'FCPS': 2,
    'MD': 3,
    'MS': 2.5,
    'PHD': 3.5
}

def score_qualifications(education):
    score = 0
    for q in qualification_scores:
        if q.lower() in str(education).lower():
            score += qualification_scores[q]
    return score

# Apply score
df['Qualification Score'] = df['Education'].apply(score_qualifications)

# Normalize features
df['Norm Experience'] = df['Experience'] / df['Experience'].max()
df['Norm Qualification'] = df['Qualification Score'] / df['Qualification Score'].max()

# Weighted score
df['Recommendation Score'] = (
    0.6 * df['Norm Experience'] +
    0.4 * df['Norm Qualification']
)

# Top doctors
top_doctors = df.sort_values("Recommendation Score", ascending=False)

print(top_doctors[['Doctor Name','Speciality','Experience','Recommendation Score']].head(10))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

df['Concentration'] = df['Concentration'].astype(str)

tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['Concentration'])

similarity = cosine_similarity(tfidf_matrix)

indices = pd.Series(df.index, index=df['Doctor Name']).drop_duplicates()

def recommend_similar_doctors(name):
    idx = indices[name]
    scores = list(enumerate(similarity[idx]))
    scores = sorted(scores, key=lambda x: x[1], reverse=True)

    scores = scores[1:6]
    doctor_indices = [i[0] for i in scores]

    return df['Doctor Name'].iloc[doctor_indices]
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import NearestNeighbors

df_knn = df.copy()

# Encode speciality
le = LabelEncoder()
df_knn['Speciality_encoded'] = le.fit_transform(df_knn['Speciality'])

features = df_knn[['Speciality_encoded','Experience']]

knn = NearestNeighbors(n_neighbors=5)
knn.fit(features)

def recommend_doctors_knn(name):

    idx = df_knn[df_knn['Doctor Name']==name].index[0]

    distances, indices = knn.kneighbors([features.iloc[idx]])

    return df_knn.iloc[indices[0][1:]]['Doctor Name']
```
