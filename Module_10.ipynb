```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression,SGDRegressor
from sklearn.metrics import mean_squared_error,r2_score
```

```python
hours=np.array([2,4,6,8,10]).reshape(-1,1)
marks=np.array([55,60,70,80,88])
hours,marks
```

```python
model=LinearRegression()
model.fit(hours,marks)
```

```python
print("m: ",model.coef_[0])
print("C: ",model.intercept_)
```

```python
plt.scatter(hours,marks,color='blue',label="Actual Data")
predicted_marks=model.predict(hours)
plt.plot(hours,predicted_marks,color='red',label='Predicted Line')
plt.xlabel('Hours')
plt.ylabel('Marks')
plt.title("Study hours vs Marks")
plt.legend()
plt.grid(True)
plt.show()
```

```python
new_hours=np.array([[7]])
prediction=model.predict(new_hours)
print("Predicted Marks",prediction[0])
```

```python
linear_reg=LinearRegression()
linear_reg.fit(hours,marks)

print("M ;",linear_reg.coef_[0])
print("C: ",linear_reg.intercept_)
```

```python
pred_marks=linear_reg.predict(hours)
mse_lin=mean_squared_error(marks,pred_marks)
print("Mean Squared Error (Linear Regression):",mse_lin)
```

GRadient descent Based model

```python
sgd_reg=SGDRegressor(
    max_iter=1000,
    learning_rate='invscaling',
    eta0=0.01,
    random_state=42)
sgd_reg.fit(hours,marks)
print("m: ",sgd_reg.coef_[0])
print("C: ",sgd_reg.intercept_[0])

pred_marks_sgd=sgd_reg.predict(hours)
mse_sgd=mean_squared_error(marks,pred_marks_sgd)
print("Mean Squared Error (SGD Regression):",mse_sgd)
```

```python
from matplotlib.lines import lineStyles
plt.scatter(hours,marks,color='blue',label="Actual Data")
predicted_marks=model.predict(hours)
plt.plot(hours,predicted_marks,color='red',label='Predicted Line',linestyle='--')
plt.plot(hours,pred_marks_sgd,color='green',label='Predicted Line (SGD)',linestyle='-.')
plt.xlabel('Hours')
plt.ylabel('Marks')
plt.title("Linear regression vs sgd")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae=mean_absolute_error(marks,pred_marks)
mse=mean_squared_error(marks,pred_marks)
r2=r2_score(marks,pred_marks)
rmse=np.sqrt(mse)
print("Mean Absolute Error:",mae)
print("Mean Squared Error:",mse)
print("Root Mean Squared Error:",rmse)
print("R-squared:",r2)
```
