# Question 01: [ Marks 10 ]

What is a perceptron?
Explain its three main components and their roles.

**Answer 01:**


A perceptron is the most basic neural network model. It is used for binary classification. It gives output like 0 or 1.

Inputs (x1, x2, …)
Data or features given to the perceptron

Weights (w1, w2, …)
Weights show how important each input is

Bias (b)
Bias helps shift the decision boundary


Perceptron formula:
y = f(w1x1 + w2x2 + b)

# Question 02: [ Marks 10 ]

What is a decision boundary in a perceptron?
Explain how the equation
w₁x₁ + w₂x₂ + b = 0
represents a decision boundary geometrically.

## Write Answer 02:

Decision boundary is a line or plane that separates two classes.

Equation:
w1x1 + w2x2 + b = 0

This equation makes a straight line in graph.

If value > 0 - Class 1


If value < 0 - Class 0

Weights control slope of line and bias controls position of line.

# Question 03: [ Marks 15 ]

Why is the perceptron called a linear classifier?
Explain your answer using the concept of linear separability and logical gate examples such as AND, OR, and XOR.

## Write Answer 03:

Perceptron is called linear classifier because it separates data using straight line.

Linear separable means data can be separated using one straight line.

Examples:

AND gate -> Can be separated -so,Perceptron works


OR gate -> Can be separated - so,Perceptron works



XOR gate -> Cannot be separated -so, Perceptron fails

So perceptron only works for linear problems.

# Question 04: [ Marks 15 ]

Write the perceptron weight update rule and explain it intuitively.
Why are the weights updated only when the prediction is incorrect?

## Write Answer 04:
Weight update rule:

w= w + η * error * x


b= b + η * error

Here,
error = Actual output − Predicted output

If prediction is wrong, weight is updated so decision boundary move to correct side.


 If prediction is correct, no need weight update.

## Write Answer 05:
