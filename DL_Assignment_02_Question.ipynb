# Question 01: [ Marks 10 ]
Explain gradient descent in detail. What is the main idea behind it? How does it help in minimizing the loss function? Explain the role of learning rate and describe what happens if the learning rate is too high or too low.

**Write** Answer 01:

Gradient Descent is a method used to reduce error in a machine learning model. Basically, the model keeps changing its weights step by step so the loss becomes smaller.

It checks the slope of the loss function. The slope shows in which direction the error is increasing. So the algorithm moves in the opposite direction to reduce the error.

Formula:

New value = Old value − Learning rate × Gradient

This process repeats many times until the loss becomes low.

Learning rate means how big the step will be.

If the learning rate is too high, the steps become too big and the model can skip the lowest point or behave unstable.

If the learning rate is too low, the steps become very small, so training becomes slow and may take a long time.

# Question 02: [ Marks 10 ]

What is forward propagation in an artificial neural network? Explain step by step how data moves from the input layer to the output layer.

## Write Answer 02:
Forward propagation is when input data goes through the network to produce output.

Steps:

Input Layer - Data enters the network

Weighted Sum - Inputs multiply with weights and adds bias

z = w1x1 + w2x2 + b

Activation Function - Adds non-linearity

a = f(z)

Next Layer - Output goes to next layer

Output Layer - Final prediction comes

In forward propagation, data only moves forward. Weights do not change.

## Write Answer 04:
Sigmoid formula:

σ(x) = 1 / (1 + e⁻ˣ)

It converts any number into a value between 0 and 1.

Where used:

-Binary classification
-Probability output

Advantages:

-Output between 0 and 1


-Smooth curve


-Easy to understand




Disadvantages:

-Learning can be slow


-Not zero-centered

Main problem:

Sigmoid causes vanishing gradient problem.
When input is very large or very small, gradient becomes tiny.
So weights update very slowly.
